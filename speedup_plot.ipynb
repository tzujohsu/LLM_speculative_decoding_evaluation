```python
import matplotlib.pyplot as plt
import numpy as np

# Define the number of variables
num_vars = 9

# Create a figure and a polar subplot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Ensure the plot is closed

# The radar chart example data
# Assume we have 4 models to compare
values_model = [[1.4, 1.50514527845036,1.09969276896596, 0.985278398940351, 1.44182327069172, 1.12288340100374,0.868019648276656,0.936235953083851,1.41410842586544],
                [1.12, 1.21565059144676, 0.919266579925993, 1.1244754774099, 1.17194388777555, 1.02958185964576,1.21277819657904,0.980866522563806,1.08249672632038], 
                [1.31,1.46402660217654, 0.887746756384995, 0.949949526855609, 1.30972388955582, 1.12831676158122, 0.820756756619551,0.950409988794065,1.25644916540212], 
                [1,1.104833836858, 0.786330122760786, 0.877667341866836, 1.04193290734824, 0.868473256601218, 0.865231744378637,0.955743926534842,0.981889591970325]]

# Define the model names
model_names = ['opt-6.7b/opt-125m', 'opt-6.7b/opt-350m', 'opt-6.7b/opt-1.3b', 'opt-6.7b/opt-2.7b']

# Draw one axe per variable + add labels labels yet
labels = ['Overall','Multi-turn','Translation','Summarization','QA', 'Mathematical Reasoning', 'RAG','Law Analtyics','Grammar Correction']

# Plot data and fill with color for each model
colors = ['lightblue','pink', 'yellow', 'orange']
for i, values in enumerate(values_model):
    values += values[:1]  # Ensure the plot is closed
    ax.plot(angles, values, color=colors[i], linewidth=2, linestyle='solid', label=model_names[i])
    ax.fill(angles, values, color=colors[i], alpha=0.15)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, color=(0, 0, 0, 0.8),fontsize=15)  # RGBA color code

# Add a title
plt.title('Performance Speedup Across Different OPT Model Sizes', size=20, color=(0, 0, 0, 0.8), y=1.1)  # RGBA color code

# Add a legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the plot
plt.show()
```
